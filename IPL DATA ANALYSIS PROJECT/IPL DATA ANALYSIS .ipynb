```python
# Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('matches.csv')
df.head()
                   
```

```python
# Check the structure of the dataset
df.info()
```

```python
df.columns
```

```python
# Get summary statistics for numerical columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
#  Handle Missing Data
# Drop columns with more than 50% missing values (if necessary)
df = df.dropna(thresh=len(df) * 0.5, axis=1)

# Drop rows with any missing data (only if feasible)
df = df.dropna()
df
```

```python

# Fill missing values in 'city' with 'Unknown'
df['city'].fillna('Unknown', inplace=True)

# Fill missing values in 'winner' with 'No Result'
df['winner'].fillna('No Result', inplace=True)

# Verify that missing values are handled
df.isnull().sum()

```

```python
# Check for duplicate rows
print(f"Duplicate rows: {df.duplicated().sum()}")

# Drop duplicate rows if found
df = df.drop_duplicates()
```

```python
# Convert Data Types
# Convert 'season' to categorical type
df['season'] = df['season'].astype('category')

# Convert 'date' to datetime type
df['date'] = pd.to_datetime(df['date'])

# Verify data type changes
df.info()
```

```python
# Standardize team names (example replacements)
team_name_mapping = {
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Kings XI Punjab': 'Punjab Kings',
    'Delhi Daredevils': 'Delhi Capitals'
}

# Apply standardization to relevant columns
for column in ['team1', 'team2', 'toss_winner', 'winner']:
    df[column] = df[column].replace(team_name_mapping)

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
# Create a new feature: 'Toss_Winner_Match_Winner' (if toss winner is also the match winner)
df['Toss_Winner_Match_Winner'] = np.where(df['toss_winner'] == df['winner'], 1, 0)

# Verify new features
df.head()
```

```python
# Save the cleaned and prepared data to a new CSV file
df.to_csv("cleaned_matches.csv", index=False)
print("Cleaned dataset saved successfully.")
```

```python
# Load the dataset
df = pd.read_csv('cleaned_matches.csv')
df.head()
                
```

```python
# Which team has the most wins?
# Count the number of wins by each team
team_wins = df['winner'].value_counts()
print("Total wins by each team:")
print(team_wins)
# Plotting total wins by each team
team_wins = df['winner'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=team_wins.index, y=team_wins.values, palette='viridis')
plt.xticks(rotation=90)
plt.title("Total Wins by Each Team")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.show()
```

```python
# What are the most common venues for matches?
# Plotting the most common venues
venue_counts = df['venue'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(y=venue_counts.index, x=venue_counts.values, palette='magma')
plt.title("Top 10 Most Common Venues")
plt.xlabel("Number of Matches")
plt.ylabel("Venue")
plt.show()
```

```python
# Who are the most frequent "Player of the Match" awardees?
# Plotting top 10 players with the most "Player of the Match" awards
player_of_match_counts = df['player_of_match'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=player_of_match_counts.index, y=player_of_match_counts.values, palette='rocket')
plt.xticks(rotation=90)
plt.title("Top 10 Players with Most 'Player of the Match' Awards")
plt.xlabel("Player")
plt.ylabel("Number of Awards")
plt.show()

```

```python
# Is there a difference in win percentages when batting vs. fielding first?
# Plotting win percentages for batting vs fielding first
batting_first_wins = df[(df['toss_decision'] == 'bat') & (df['toss_winner'] == df['winner'])].shape[0]
fielding_first_wins = df[(df['toss_decision'] == 'field') & (df['toss_winner'] == df['winner'])].shape[0]
total_batting_first = df[df['toss_decision'] == 'bat'].shape[0]
total_fielding_first = df[df['toss_decision'] == 'field'].shape[0]
batting_win_percentage = (batting_first_wins / total_batting_first) * 100
fielding_win_percentage = (fielding_first_wins / total_fielding_first) * 100

plt.figure(figsize=(8, 4))
sns.barplot(x=['Batting First', 'Fielding First'], y=[batting_win_percentage, fielding_win_percentage], palette='Set2')
plt.title("Win Percentage When Batting vs Fielding First")
plt.xlabel("Decision")
plt.ylabel("Win Percentage")
plt.show()
```

```python
#  How does the win margin vary by team?
# Plotting average win margin by team
win_margin_avg = df.groupby('winner')['result_margin'].mean().sort_values(ascending=False).dropna()
plt.figure(figsize=(12, 6))
sns.barplot(x=win_margin_avg.index, y=win_margin_avg.values, palette='Blues')
plt.xticks(rotation=90)
plt.title("Average Win Margin by Team")
plt.xlabel("Team")
plt.ylabel("Average Win Margin (Runs/Wickets)")
plt.show()
```

```python
# Has the average score required to win changed over the years?
# Plotting average target runs per season
average_target_by_season = df.groupby('season')['target_runs'].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x=average_target_by_season.index, y=average_target_by_season.values, marker="o", color='purple')
plt.title("Average Target Runs Required to Win by Season")
plt.xlabel("Season")
plt.ylabel("Average Target Runs")
plt.show()
```

```python
# Top 5 Most Frequent Team Rivalries
rivalries = df.groupby(['team1', 'team2']).size().nlargest(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=rivalries.values, y=[f"{t1} vs {t2}" for t1, t2 in rivalries.index], palette="viridis")
plt.title("Top 5 Most Frequent Team Rivalries")
plt.xlabel("Number of Matches")
plt.ylabel("Rivalry")
plt.show()

```

```python
# Predicting Match Winner Based on Team and Venue Information
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Selecting features and target
features = df[['team1', 'team2', 'venue', 'toss_decision']]
target = df['winner']

# Encode categorical data
features = pd.get_dummies(features, drop_first=True)
target = target.astype('category').cat.codes  # Convert target to numerical codes

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

print(y_pred)
```

```python
# Predicting if Toss Decision Affects Match Outcome
# Preparing features for toss-related prediction
toss_features = df[['toss_winner', 'toss_decision', 'season']]
toss_target = (df['toss_winner'] == df['winner']).astype(int)  # 1 if toss winner wins the match, else 0

# Encode categorical data
toss_features = pd.get_dummies(toss_features, drop_first=True)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(toss_features, toss_target, test_size=0.3, random_state=42)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(y_pred)
print(f"Model Accuracy for Toss Impact: {accuracy:.2f}")
```
